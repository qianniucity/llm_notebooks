````python
import os
from groq import Groq
from IPython.display import display, Markdown
import time

groq_api_key = os.environ["GROQ_API_KEY"]

client = Groq(api_key=groq_api_key)
start_time = time. time() # å¼€å§‹è®¡æ—¶
chat_completion = client.chat.completions.create(
    messages=[
        {
            "role": "system",
            "content":"""
            - Think step by step.
            - Be precise, no preamble, get to the point.
            - Always response in Simplified Chinese, unless user is in Eng Lish.
            """
        },{
            "role": "user",
            "content":"ä½ æ˜¯ä¸€ä½ä¸“ä¸šçš„ç»æµå­¦å®¶ï¼Œè¯·ç”¨å°ç™½èƒ½ç†è§£çš„è¡¨è¾¾æ¥è¯¦ç»†è§£ææ²‰æ²¡æˆæœ¬ï¼Œè¾“å‡ºä¸º Markdown æ ¼å¼ï¼Œå¤§æ ‡é¢˜å‰æ–¹è¯·é™„ä¸Š 1 ä¸ªåˆé€‚çš„ Emoji",
        }
        ],
    model="llama3-70b-8192",
    temperature=0.8, max_tokens=4096,top_p=0.9,
)

end_time = time.time() # ç»“æŸè®¡æ—¶
elapsed_time = end_time - start_time # è®¡ç®—ç”Ÿæˆæ—¶é—´
generated_chars = len(chat_completion.choices[0].message.content) # è·å–ç”Ÿæˆçš„å­—ç¬¦æ•°
speed = generated_chars / elapsed_time # è®¡ç®—ç”Ÿæˆé€Ÿåº¦

display(Markdown(f"ç”Ÿæˆé€Ÿåº¦:{speed:.0f}å­—ç¬¦/ç§’"))
display(Markdown(chat_completion.choices[0].message.content))
````
